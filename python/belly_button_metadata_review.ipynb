### Rename Column Headers

```python
df = df.rename(columns={"EVENT":"event", "ETHNICITY":"ethnicity","GENDER":"gender","AGE":"age","WFREQ":"wfreq","BBTYPE":"bbtype","LOCATION":"location","COUNTRY012":"country_012"})
df.head()                 
```

```python
df = df.rename(columns={"ZIP012":"zip_012", "COUNTRY1319":"country_1319","ZIP1319":"zip_1319","DOG":"dog","CAT":"cat","IMPSURFACE013":"impsurface_013","NPP013":"npp_013","MMAXTEMP013":"maxtemp_013"})
df.head()   
```

```python
df = df.rename(columns={"PFC013":"pfc_013", "IMPSURFACE1319":"impsurface_1319","NPP1319":"npp_1319","MMAXTEMP1319":"mmaxtemp_1319","PFC1319":"pfc_1319"})
df.head() 
```

```python
list_of_columns = ["age","wfreq","zip_012","zip_1319","impsurface_013","npp_013","maxtemp_013","impsurface_1319","npp_1319","mmaxtemp_1319","pfc_1319"]
```

```python
i = 0

while i <= len(list_of_columns):
    for cols in list_of_columns: 
        df[cols] = df[cols].fillna(0)
    i = i + 1
```

```python
df.head()
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df['dog'] = df['dog'].fillna('none')
df['cat'] = df['cat'].fillna('none')
```

```python
df.dropna(subset=['event'], inplace=True)
```

```python
df.head()
```

```python
list_of_cols_upper = ["gender","bbtype","country_012","country_1319"]
```

```python
i = 0

while i <= len(list_of_cols_upper):
    for cols in list_of_cols_upper: 
        df[cols] = df[cols].str.upper()
    i = i + 1
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['impsurface_1319'] = df['impsurface_1319'].replace("'NA'",0).astype('float64')
```

```python
df.dtypes
```

```python
df.to_csv('../db/belly_button_summarized.csv', sep=',' , encoding= 'utf-8', index=False)
```

```python
## the bacteria have the same prefix, thus is in order in to get the unique lables for each sample
## I've extracted the last character after ";" to indentify the bacteria, rather than the whole name itself. 
## for readability 

i= 0
button_columns = []

for row in df_belly["otu_label"]: 
    button_columns.append(row.split(";")[-1])
    
    i = i + 1 
    
    
print(button_columns)
```

```python
#inserted column of otu summary labels as the first real daata column to eventually act as primary index value for charts

df_belly.insert(loc=1, column='otu_summary_label', value=button_columns)
df_belly.head()
```

```python
# to drop csv index value for rows and descriptive otu label
df_belly_button_data = df_belly.drop(columns = ["otu_id","otu_label"])
```

```python
df_belly_button_data.head()
```

```python
#sort summary labels in asc order
df_belly_button_data.sort_values(by="otu_summary_label", ascending=True)
```

```python
# to create a list of sample columns for groupby function
sample_cols = list(df_belly_button_data.iloc[:,1:].columns)
```

```python
# to check the list of columns length
len(sample_cols)
```

```python
# to make sure that the list of sample cols do not have duplicate items
len(set(sample_cols))
```

```python
# adjust dataframe to be a groupby summary of the summarized label per sample size
belly_button_df = df_belly_button_data.groupby('otu_summary_label')[sample_cols].sum().reset_index()
```

```python
belly_button_df.head()
```

```python
engine = create_engine ('postgresql://postgres:postgres@localhost:5432/project_databases')
belly_button_df.to_sql('belly_button_metadata', engine)
```

```python
belly_button_df.to_csv('../db/belly_button_metadata_summarized.csv', sep=',' , encoding= 'utf-8', index=False)
```
